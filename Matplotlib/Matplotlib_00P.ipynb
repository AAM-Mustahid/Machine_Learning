### Matplotlib Figure Object

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = 2 * x
```

## Creating figure
The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.5,0.5]) # left, bottom, width, height (range 0 to 1)
plt.plot(x,y)
plt.plot(a,b)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title("oAw")
plt.show()
```

### Adding another set of axes to the Figure
we've only seen one set of axes on this figure object, but we can keep adding new axes on to it at any location and size we want. We can then plot on that new set of axes.

```python
#we plot a subplot without using subplot() function
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
sub_axes = fig.add_axes([0.2,0.5,0.2,0.2])
axes.plot(x,y)
sub_axes.plot(a,b)
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-Axis")
sub_axes.set_xlabel("sub_X")
sub_axes.set_ylabel('Sub_y')
plt.show()
```

### Some Figure parameter

```python
fig = plt.figure(figsize=(5,5))#dpi can also be use
axes = fig.add_axes([0,0,0.5,0.5])
axes.plot(a,b)
axes.set_xlabel('X_axis')
axes.set_ylabel('Y_Axis')
axes.set_title('Demo plot')

#we can save any matplotlib figure()
fig.savefig('trial_figure.png',bbox_inches = 'tight') #it supports all image format
```

## Subplots using subplot

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig,axes = plt.subplots(nrows= 1,ncols=2,figsize=(5,3))
axes[0].plot(x,y)
axes[1].plot(a,b)
axes[0].set_xlabel('X_label')
axes[0].set_ylabel('Y_label')
axes[0].set_title("Wave")
axes[1].set_xlabel('X_label')
axes[1].set_ylabel('Y_label')
axes[1].set_title("Wave")
fig.suptitle('Demo Works',fontsize = 13)
plt.show()
```

```python
fig,((axes1,axes2),(axes3,axes4)) = plt.subplots(nrows=2,ncols=2,figsize=(6,4))

axes1.plot(a,b)
axes2.plot(x,y)
axes3.plot(y,x)
axes4.plot(b,a)

fig.subplots_adjust(wspace=0.3,hspace=0.4)
plt.show()
```

## Matplotlib Styling


```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure(figsize=(7,4))
axis = fig.add_axes([0,0,0.5,0.5])

axis.plot(x, x**2, label="x**2")
axis.plot(x, x**3, label="x**3")
# Lots of options....

#ax.legend(loc=1) # upper right corner
#ax.legend(loc=2) # upper left corner
#ax.legend(loc=3) # lower left corner
#ax.legend(loc=4) # lower right corner

axis.legend(loc='upper left')#for automatic loc value is 0(default)
fig.show()
```

```python
#we can manually set value for legend loc
axis.legend(loc=(1.1,0.3))
fig
```

### Setting colors,linewidth,linetypes


Colors with matlab like syntax

```python
axis.plot(x, x**2, 'b.-') #blue line with dot
axis.plot(x, x**3, label="x**3"'g--') #green line with dashed line
fig
```

## Suggestive approach: Use Keyword arguments
color with the color parameter

```python
fig,ax = plt.subplots()
ax.plot(x,x+1,color = 'blue',alpha = 0.5,label = 'Hijibiji')
ax.plot(x,x+2,color = 'green',alpha = 1.0)
ax.plot(x,x+3,color = 'red',alpha = 0.8)
ax.set_title('Hi')
```

### Linewidth

```python
fig,ax = plt.subplots()
ax.plot(x,x+1,color = 'blue',alpha = 0.5,label = 'Hijibiji')
ax.plot(x,x+2,color = 'green',alpha = 1.0,linewidth = 0.5,linestyle = '--')
ax.plot(x,x+3,color = 'red',alpha = 0.8)
ax.set_title('Hi')
```

you can do custom plot in matplotlib

```python
fig,ax = plt.subplots()
ax.plot(x,x+1,color = 'blue',alpha = 0.5,marker = 'p',ms = 10,label = 'Hijibiji')
ax.plot(x,x+2,color = 'green',alpha = 1.0)
ax.plot(x,x+3,color = 'red',alpha = 0.8)
ax.set_title('Hi')
```
