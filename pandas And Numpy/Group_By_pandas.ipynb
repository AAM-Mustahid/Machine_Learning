### Grab elements based on cross section with .xs()
##### this method takes a key to select data at a particular level of a multi-index

```python
#first we grab data without cross section
test = df[df['cylinders'].isin([8])].groupby(['model_year',"cylinders"]).mean()
```

```python
#by using cross section
moodel.xs(key=8,axis=0,level=1)
```

```python
moodel
```

```python
moodel.index
```

##### Sorting Multi Index
